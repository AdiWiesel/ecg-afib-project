```python

import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from typing import Tuple
import tensorflow as tf
from tensorflow.keras import layers, callbacks, Model
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score, roc_curve,
    precision_recall_curve, PrecisionRecallDisplay
)


DATA_PATH   = "mitdb_windows_5s_binary_afib.npz"  # require X, y, rec_ids
RANDOM_SEED = 42
BATCH_SIZE  = 256
EPOCHS      = 12
LR          = 1e-3
DOWNSAMPLE  = 3          # 360 Hz -> 120 Hz (1800 -> 600)
BASE_FILTER = 16         

# Threshold objective: 
THR_MODE    = "f1"
THR_TARGET  = 0.85       # used if mode is precision_at or recall_at

SHOW_PLOTS  = True


rng = np.random.default_rng(RANDOM_SEED)

def summarize_split(name, yv):
    n = len(yv); af = int(yv.sum())
    print(f"{name}: {n} | AFib {af} ({af/n:.2%}), Normal {n-af} ({1-af/n:.2%})")

def split_list(lst, ratios=(0.7, 0.15, 0.15), seed=RANDOM_SEED):
    lst = lst.copy()
    rng = np.random.default_rng(seed)
    rng.shuffle(lst)
    n = len(lst); a = int(round(ratios[0]*n)); b = int(round(ratios[1]*n))
    return lst[:a], lst[a:a+b], lst[a+b:]

def choose_threshold(y_true, probs, mode="f1", target=0.85) -> Tuple[float, dict]:
    """Return threshold and a dict of summary metrics on validation."""
    precision, recall, thresholds = precision_recall_curve(y_true, probs)
    # thresholds has len = len(precision)-1
    f1 = (2 * precision * recall) / (precision + recall + 1e-12)
    best_thr = 0.5
    info = {}

    if mode == "f1":
        idx = np.nanargmax(f1[:-1])  # exclude last point where threshold is undefined
        best_thr = thresholds[idx]
        info = {"precision": float(precision[idx]), "recall": float(recall[idx]), "f1": float(f1[idx])}

    elif mode == "precision_at":
        # pick smallest thr that achieves precision >= target (highest recall among those)
        mask = precision[:-1] >= target
        if np.any(mask):
            # among acceptable precision, choose the one with highest recall
            idxs = np.where(mask)[0]
            pick = idxs[np.argmax(recall[idxs])]
            best_thr = thresholds[pick]
            info = {"precision": float(precision[pick]), "recall": float(recall[pick]), "f1": float(f1[pick])}
        else:
            # fallback to max F1
            idx = np.nanargmax(f1[:-1])
            best_thr = thresholds[idx]
            info = {"precision": float(precision[idx]), "recall": float(recall[idx]), "f1": float(f1[idx]), "note": "precision target not reachable; used max F1"}

    elif mode == "recall_at":
        # pick largest thr that achieves recall >= target (to improve precision)
        mask = recall[:-1] >= target
        if np.any(mask):
            idxs = np.where(mask)[0]
            # among acceptable recall, choose highest precision (which will typically be at higher thr)
            pick = idxs[np.argmax(precision[idxs])]
            best_thr = thresholds[pick]
            info = {"precision": float(precision[pick]), "recall": float(recall[pick]), "f1": float(f1[pick])}
        else:
            idx = np.nanargmax(f1[:-1])
            best_thr = thresholds[idx]
            info = {"precision": float(precision[idx]), "recall": float(recall[idx]), "f1": float(f1[idx]), "note": "recall target not reachable; used max F1"}
    else:
        raise ValueError("THR_MODE must be one of {'f1','precision_at','recall_at'}")

    return float(best_thr), info

# data
d = np.load(DATA_PATH, allow_pickle=True)
assert {"X","y","rec_ids"}.issubset(set(d.files)), "Dataset must contain X, y, rec_ids"
X, y, rec_ids = d["X"], d["y"].astype(int), d["rec_ids"].astype(str)
print("Loaded:", X.shape, y.shape, rec_ids.shape)

# Downsample 
X = X[:, ::DOWNSAMPLE]
X = X[..., None]  # (N, 600, 1)
print("Downsampled shape:", X.shape)

# Grouped + stratified split by record (ensure AFib records in each split)
rec_to_idx = defaultdict(list)
for i, r in enumerate(rec_ids): rec_to_idx[r].append(i)
records = sorted(rec_to_idx.keys())
rec_has_afib = {r: (y[rec_to_idx[r]].mean() > 0.0) for r in records}
recs_pos = [r for r in records if rec_has_afib[r]]
recs_neg = [r for r in records if not rec_has_afib[r]]

pos_tr, pos_va, pos_te = split_list(recs_pos, (0.7, 0.15, 0.15), RANDOM_SEED)
neg_tr, neg_va, neg_te = split_list(recs_neg, (0.7, 0.15, 0.15), RANDOM_SEED+1)

train_recs = set(pos_tr + neg_tr)
val_recs   = set(pos_va + neg_va)
test_recs  = set(pos_te + neg_te)

def idx_from_recs(rec_set):
    idxs = []
    for r in rec_set: idxs.extend(rec_to_idx[r])
    return np.array(sorted(idxs), dtype=int)

train_idx = idx_from_recs(train_recs)
val_idx   = idx_from_recs(val_recs)
test_idx  = idx_from_recs(test_recs)

X_train, y_train = X[train_idx], y[train_idx]
X_val,   y_val   = X[val_idx],   y[val_idx]
X_test,  y_test  = X[test_idx],  y[test_idx]

print("\n=== Split summary (grouped & stratified) ===")
summarize_split("Train", y_train)
summarize_split("Val",   y_val)
summarize_split("Test",  y_test)

# datasets :
def make_balanced_train_ds(X, y, batch_size=BATCH_SIZE, seed=RANDOM_SEED):
    pos = y == 1
    Xp, yp = X[pos], y[pos]
    Xn, yn = X[~pos], y[~pos]
    ds_pos = tf.data.Dataset.from_tensor_slices((Xp, yp)).shuffle(len(Xp), seed=seed, reshuffle_each_iteration=True).repeat()
    ds_neg = tf.data.Dataset.from_tensor_slices((Xn, yn)).shuffle(len(Xn), seed=seed+1, reshuffle_each_iteration=True).repeat()
    ds = tf.data.Dataset.sample_from_datasets([ds_pos, ds_neg], weights=[0.5, 0.5], seed=seed)
    return ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)

def make_eval_ds(X, y, batch_size=BATCH_SIZE):
    return tf.data.Dataset.from_tensor_slices((X, y)).batch(batch_size).prefetch(tf.data.AUTOTUNE)

ds_train = make_balanced_train_ds(X_train, y_train)
ds_val   = make_eval_ds(X_val,   y_val)
ds_test  = make_eval_ds(X_test,  y_test)

#model- (1Dconv)
@tf.function
def binary_focal_loss(y_true, y_pred, gamma=2.0, alpha=0.30, eps=1e-7):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.clip_by_value(y_pred, eps, 1.0 - eps)
    pt = tf.where(tf.equal(y_true, 1.0), y_pred, 1.0 - y_pred)
    w  = tf.where(tf.equal(y_true, 1.0), alpha, 1.0 - alpha)
    return -tf.reduce_mean(w * tf.pow(1.0 - pt, gamma) * tf.math.log(pt))

def tiny_cnn(input_len=600, base=BASE_FILTER):
    inp = layers.Input(shape=(input_len, 1), name="ecg")
    x = layers.SeparableConv1D(base,   11, strides=2, padding="same")(inp); x = layers.BatchNormalization()(x); x = layers.ReLU()(x)  # /2
    x = layers.SeparableConv1D(base,    7, strides=2, padding="same")(x);   x = layers.BatchNormalization()(x); x = layers.ReLU()(x)  # /4
    x = layers.SeparableConv1D(base*2,  5, strides=2, padding="same")(x);   x = layers.BatchNormalization()(x); x = layers.ReLU()(x)  # /8
    x = layers.SeparableConv1D(base*2,  5, strides=2, padding="same")(x);   x = layers.BatchNormalization()(x); x = layers.ReLU()(x)  # /16
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(32, activation="relu")(x); x = layers.Dropout(0.20)(x)
    out = layers.Dense(1, activation="sigmoid", name="afib_prob")(x)
    return Model(inp, out)

model = tiny_cnn(input_len=X_train.shape[1])
model.compile(optimizer=tf.keras.optimizers.Adam(LR),
              loss=binary_focal_loss,
              metrics=[tf.keras.metrics.BinaryAccuracy(name="acc"),
                       tf.keras.metrics.AUC(name="auc"),
                       tf.keras.metrics.Precision(name="prec"),
                       tf.keras.metrics.Recall(name="rec")])
model.summary()
# early stoppage
early   = callbacks.EarlyStopping(monitor="val_auc", mode="max", patience=3, restore_best_weights=True)
plateau = callbacks.ReduceLROnPlateau(monitor="val_auc", mode="max", factor=0.5, patience=1, min_lr=1e-5, verbose=1)

steps_per_epoch = max(100, len(y_train)//BATCH_SIZE)
hist = model.fit(ds_train, validation_data=ds_val,
                 steps_per_epoch=steps_per_epoch, epochs=EPOCHS,
                 callbacks=[early, plateau], verbose=1)


val_probs = model.predict(ds_val, verbose=0).ravel()
val_auc   = roc_auc_score(y_val, val_probs)
print(f"\nValidation AUC = {val_auc:.3f}")

thr, val_info = choose_threshold(y_val, val_probs, mode=THR_MODE, target=THR_TARGET)
print(f"Chosen threshold (mode='{THR_MODE}', target={THR_TARGET}): {thr:.3f}")
print(f"Val @thr → precision={val_info.get('precision',np.nan):.3f}, recall={val_info.get('recall',np.nan):.3f}, f1={val_info.get('f1',np.nan):.3f}{' | '+val_info.get('note','') if 'note' in val_info else ''}")

# test
test_probs = model.predict(ds_test, verbose=0).ravel()
test_auc   = roc_auc_score(y_test, test_probs)
print(f"\nTest AUC = {test_auc:.3f}")

def eval_at_threshold(y_true, probs, thr, title):
    pred = (probs >= thr).astype(int)
    print(f"\n=== {title} (thr={thr:.3f}) ===")
    print(classification_report(y_true, pred, target_names=["Normal","AFib"]))
    print("Confusion Matrix:\n", confusion_matrix(y_true, pred))

eval_at_threshold(y_test, test_probs, thr, "Test (tuned)")
eval_at_threshold(y_test, test_probs, 0.5, "Test (0.5)")

# plot :
if SHOW_PLOTS:
    # Training curves
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1); plt.plot(hist.history["loss"]);    plt.plot(hist.history["val_loss"]); plt.title("Loss"); plt.legend(["train","val"]); plt.grid(True)
    plt.subplot(1,2,2); plt.plot(hist.history["auc"]);     plt.plot(hist.history["val_auc"]);  plt.title("AUC");  plt.legend(["train","val"]); plt.grid(True)
    plt.tight_layout(); plt.show()

    # ROC
    fpr, tpr, _ = roc_curve(y_test, test_probs)
    plt.figure(figsize=(5,5))
    plt.plot(fpr, tpr, label=f"AUC={test_auc:.2f}")
    plt.plot([0,1],[0,1],'k--'); plt.xlabel("FPR"); plt.ylabel("TPR")
    plt.title("ROC - Test (Grouped & Stratified)"); plt.legend(); plt.grid(True); plt.tight_layout(); plt.show()

    # Precision-Recall (Val) + chosen threshold marker
    precision, recall, thresholds = precision_recall_curve(y_val, val_probs)
    disp = PrecisionRecallDisplay(precision=precision, recall=recall)
    disp.plot()
    # Show point closest to chosen threshold on PR curve
    if len(thresholds) > 0:
        idx_thr = np.argmin(np.abs(thresholds - thr))
        plt.scatter(recall[idx_thr], precision[idx_thr], s=60, c="red", label=f"thr={thr:.2f}")
        plt.legend()
    plt.title("Precision–Recall (Validation)"); plt.grid(True); plt.tight_layout(); plt.show()

    # Probability histograms (Test)
    plt.figure(figsize=(6,4))
    plt.hist(test_probs[y_test==0], bins=40, alpha=0.6, label="Normal")
    plt.hist(test_probs[y_test==1], bins=40, alpha=0.6, label="AFib")
    plt.axvline(thr, color="red", linestyle="--", label=f"thr={thr:.2f}")
    plt.title("Test probabilities"); plt.legend(); plt.tight_layout(); plt.show()
```
